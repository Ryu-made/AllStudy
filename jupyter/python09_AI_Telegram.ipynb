# 파이썬의 비동기처리

```python
import asyncio # 주피터랩 등에서는 충돌 가능성
import nest_asyncio # 개선된 모듈
import time
```

```python
def task1():
    print('작업 1 시작')
    time.sleep(2.1)
    print('작업 1 완료')
def task2():
    print('작업 2 시작')
    time.sleep(2)
    print('작업 2 완료')
task1()
task2()
```

```python
nest_asyncio.apply()
async def atask1():
    print('작업 1 시작')
    await asyncio.sleep(1.25)
    print('작업 1 완료')
async def atask2():
    print('작업 2 시작')
    await asyncio.sleep(0.75)
    print('작업 2 완료')

async def main():
    asyncio.gather(atask1(), atask2()) # 두 작업을 동시에 실행

asyncio.run(main())
```
