```python
from konlpy.tag import Okt
from collections import Counter

import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib

import pandas as pd
import numpy as np
```

```python
plt.style.use('fivethirtyeight')

# D2Coding 폰트 설정
d2coding_path = './D2Coding-Ver1.3.2-20180524.ttf' # 폰트 경로
fm.fontManager.addfont(d2coding_path)
plt.rcParams['font.family'] = 'D2Coding'

# 마이너스 부호 깨지지 않도록 설정
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
def extract_sentiment_words(okt, text, pos_dc, neg_dc):
    # 형태소 분석 후 어휘 추출(Okt.morphs)
    words = okt.morphs(text, stem=True)
    # 긍정 부정 단어 분류
    posi = [w for w in words if w in pos_dc]
    nega = [w for w in words if w in neg_dc]
    return posi, nega
```

```python
text = '''
이 제품은 정말 훌륭하고, 만족스러워요. 배송이 느려서 좀 아쉬워요. 
서비스는 친절하고, 가격은 너무 비싸네요. 마음이 불편해요
'''

# pos_wd = []
# neg_wd = []

pos_wd = ['훌륭하다', '만족스럽다', '친절하다', '기쁘다', '편리하다', '좋다', '고맙다', '행복하다', '아름답다']
neg_wd = ['느리다', '열받다', '비싸다', '싫다', '골때리다', '답답하다', '시끄럽다', '아쉽다', '불편하다']


okt = Okt()
# okt.morphs(text, stem=True)

posi, nega = extract_sentiment_words(okt, text=text, pos_dc=pos_wd, neg_dc=neg_wd)


print(f'positive words: {Counter(posi)}')
print(f'negative words: {Counter(nega)}')
```

## 시각화 하기

```python
def plot_sentiment(pos, neg):
    senti_counts = {'Positive': len(pos), 'Negative': len(neg)}
    plt.bar(x=senti_counts.keys(), height=senti_counts.values(), width=0.5)
    plt.title('Sentiment Words Analysis')
    plt.ylabel('word count')
    plt.savefig('image/sentiment_analy.png')
    plt.show()


plot_sentiment(posi, nega)
```
