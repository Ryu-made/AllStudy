```python
from hdfs import InsecureClient
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib
import pandas as pd
from prophet import Prophet
import numpy as np
```

```python
plt.style.use('fivethirtyeight')

d2coding_path = 'D2Coding/D2Coding-Ver1.3.2-20180524.ttf' # 폰트 경로
fm.fontManager.addfont(d2coding_path)
plt.rcParams['font.family'] = 'D2Coding'

matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
client_hdfs = InsecureClient('http://namenode:9870', user='hadoop')
```

```python
path = '/user/hadoop/연령별취업자.csv'
```

```python
with client_hdfs.read(path, encoding='utf-8') as reader:
    data = pd.read_csv(reader)
```

# 연도별 남녀 취업자수 관계

```python
data.columns
```

```python
data.columns = ["quarter", "region", "gender", "age", "value"]
```

```python
# 데이터프레임 스타일 설정
def style_data_table(data):
    style_data = data.head(10).style.set_table_styles([
        {
            'selector': 'th, td', 
            'props': [('text-align', 'center'), ('border', '1px solid #ccc')]
        }
    ])
    return style_data
```

```python
style_data_table(data)
```

```python
data_null = data[data['value'] == '-']
data_null = data_null.groupby('region').sum().reset_index()
data_null[data_null['region'] != '세종특별자치시'][['region', 'quarter', 'value']]
```

```python
ex_data = data.copy() # 깊은 복사
ex_data['value'] = pd.to_numeric(ex_data['value'], errors='coerce')
```

```python
ex1_data = ex_data.groupby('quarter')['value'].sum().reset_index(name='total')

ax = ex1_data.plot(
    x='quarter',
    y='total',
    figsize=(15, 6),
    title='분기별 취업자수'
)
ax.set_xticks(range(len(ex1_data)))
ax.set_xticklabels(ex1_data['quarter'], rotation=45)
plt.show()
style_data_table(ex1_data)
```

```python
ex2_data = ex_data.groupby(['quarter', 'gender'])['value'].sum().reset_index(name='total')
ex2_data_gentleman = ex2_data[ex2_data['gender']=='남자'][['quarter', 'total']]
ex2_data_lady = ex2_data[ex2_data['gender']=='여자'][['quarter', 'total']]
```

```python
style_data_table(ex2_data_gentleman)
```

```python
ax_g = ex2_data_gentleman.plot(
    x='quarter',
    y='total',
    figsize=(15, 6),
    title='분기별 남성 취업자수'
)
ax_g.set_xticks(range(len(ex2_data_gentleman)))
ax_g.set_xticklabels(ex2_data_gentleman['quarter'], rotation=45)
plt.show()
style_data_table(ex2_data_gentleman)
```

```python
ax_l = ex2_data_lady.plot(
    x='quarter',
    y='total',
    figsize=(15, 6),
    title='분기별 여성 취업자수'
)
ax_l.set_xticks(range(len(ex2_data_lady)))
ax_l.set_xticklabels(ex2_data_lady['quarter'], rotation=45)
plt.show()
style_data_table(ex2_data_lady)
```

```python
# 보류
ex2_data_mixed = pd.DataFrame()
ex2_data_mixed['quarter'] = ex1_data['quarter']
ex2_data_mixed['gentleman'] = ex2_data_gentleman['total']
ex2_data_mixed['lady'] = ex2_data_lady['total']
```

```python
fig, ax = plt.subplots(figsize=(15, 6))
ax.plot(
    ex2_data_gentleman['quarter'],
    ex2_data_gentleman['total'],
    color='blue',
    marker='o'
)
ax.plot(
    ex2_data_lady['quarter'],
    ex2_data_lady['total'],
    color='red',
    marker='o'
)
ax.set_xticks(range(len(ex2_data_lady)))
ax.set_xticklabels(ex2_data_lady['quarter'], rotation=45)
ax.set_title('분기별 남여 취업자수')
ax.set_xlabel('년도')
ax.set_ylabel('인구')
plt.show()
```

```python
ex3_data_20s = ex3_data[ex3_data['age'] == '20 - 29세']
ex3_data_60s = ex3_data[ex3_data['age'] == '60세이상']
```

```python
ex3_data_20s_g = ex3_data_20s.groupby('region')['value'].sum()
ex3_data_60s_g = ex3_data_60s.groupby('region')['value'].sum()
```

```python
regions = list(ex3_data_20s_g.index)
color = [ np.random.rand(3,) for _ in range(len(ex3_data_20s_g.index)) ]
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

# 데이터
x = ex3_data_20s_g
y = ex3_data_60s_g
z = ex3_data.groupby('region')['value'].sum()
scatter = ax.scatter(x, y, s=z*0.875, c=color, alpha=0.5)
# 라벨 추가
for i in range(len(x)):
    # 지역명 붙이기
    ax.annotate(regions[i],xy=(x.iloc[i], y.iloc[i]), ha='center')
# 대각선 기준선 추가
min_val = min(min(x), min(y))
max_val = max(max(x), max(y))
ax.plot([min_val, max_val], [min_val, max_val], color='blue', linestyle='--', linewidth=2)
ax.set_title('2018년 전국 청년, 60세 이상 취업자수 관계', size=20)
ax.set_xlabel('청년 취업자')
ax.set_ylabel('60세 이상 취업자')
# 그래프 이미지 저장
plt.savefig('image/2018년20대_60대취업자수관계.png')
plt.show()
```
